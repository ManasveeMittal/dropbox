```python
file_path = '/home/manavee/Dropbox/Python/AXA/BagOfWords/docword.kos.txt'
col_names = ['docId', 'wordId', 'count']
D = pd.read_csv(file_path, skiprows=3, sep = ' ', header=None, names = col_names)#, nrows = 5000)

print( D.head(7))
print( len(D))


'''

#Hierarchical clutering upto level 2 to find tructure among the document...
A = AgglomerativeClustering(n_clusters = 2, affinity = 'cosine', linkage = 'average').fit(P)
print( A
'''
```

```python
T = pd.DataFrame(D.groupby(['wordId']).size(), columns = ['freq']).reset_index()
print( T.head(7))
print( T.groupby(['freq']).size())

```

```python
tf = 1 + np.log(pd.DataFrame(D.groupby(['wordId'])['count'].agg('sum')))
idf = pd.DataFrame(np.log(D['docId'].nunique() / (1 + ( D.groupby(['wordId']).count())))).drop('docId', axis = 1)
print( 'tf:')
print( tf.head(7))
print( 'idf:')
print( idf.head(7))
```

```python


J = tf.merge(idf, left_index = True, right_index= True)
J['tf_idf'] = J['count_x'] * J['count_y']

J.drop(['count_x', 'count_y'], axis = 1, inplace = True)
J = J.reset_index()
print( 'J:')
print( J.head(7))
print( D.head(4))

```

```python
M = D.merge(J)#, left_on = 'wordId', right_on = 'wordId')#on_leftwordId = right.wordId)
print( M.head(13))

#hist(M['tf_idf'])


P = pd.pivot_table(M, values = ['tf_idf'], index = ['docId'], columns = ['wordId']).fillna(0) #agg func mean
print( P.head(7))

#Distance matrix to etimate doc similarities...	
S = cosine_distances(P)
print( S.shape)
print( S)
```

```python
plt.imshow(S, cmap='hot', interpolation='nearest')
"""
nterpolation : string, optional, default: None
    Acceptable values are 'none', 'nearest', 'bilinear', 'bicubic',
    'spline16', 'spline36', 'hanning', 'hamming', 'hermite', 'kaiser',
    'quadric', 'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc',
    'lanczos'

    If `interpolation` is None, default to rc `image.interpolation`.
    See also the `filternorm` and `filterrad` parameters.
    If `interpolation` is 'none', then no interpolation is performed
    on the Agg, ps and pdf backends. Other backends will fall back to
    'nearest'.
    """
plt.show()
```

```python
plt.figure()

Z = linkage(P, 'ward')
dendrogram(Z, truncate_mode='lastp', p=100)
"""
truncate_mode : str, optional
    The dendrogram can be hard to read when the original
    observation matrix from which the linkage is derived is
    large. Truncation is used to condense the dendrogram. There
    are several modes:

    ``None/'none'``
      No truncation is performed (Default).

    ``'lastp'``
      The last ``p`` non-singleton formed in the linkage are the only
      non-leaf nodes in the linkage; they correspond to rows
      ``Z[n-p-2:end]`` in ``Z``. All other non-singleton clusters are
      contracted into leaf nodes.
      """

plt.show()
```
